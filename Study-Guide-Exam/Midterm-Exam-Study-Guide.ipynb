```python
#Column subsetting!

import pandas as pd

employee = pd.DataFrame({
    "Name": ["Jame", "Kate", "Omar", "Mei", "Luis"],
    "Age": [19, 24, 21, 28, 32],
    "City": ["Austin", "Riverside", "San Jose" , pd.NA, "Portland"],  # blank city as pd.NA
    "HoursWorked": [40, 45, 35, 50, 30],
})

print(employee)

print("*******************")

#all return the same result ; single column subsetting
print(employee[["Name"]])
print(employee['Name'])
print(employee.Name)

# Multiple column subseeting

print(employee[['Name', 'City']])
```

```python
employee  #missing values --> either NaN or <NA>
```

```python
# Row subsetting

#Return the second and third rows
print(employee.iloc[1:3])  #integer based

print('************')
print(employee.loc[1:2]) #label based


#first three rows and and the columns, Name and City
print("*********")
print(employee.loc[0:2, ['Name', 'City']])
print("*********")
print(employee.iloc[0:3, [0,2]])
```

```python
# advance row subsetting  -> Show me the employees whose ages are greater than 20 with two columns, Name and Age

employee[employee.Age > 20][['Name','Age']] # Note you must use double brakets --> Why?  multiple columns are used
```

```python
# Remove rows with missing values such as NaN, <NA>, etc. --> dropna

#Remove the rows with the missing values permenantly if both Name and Age columns contain NaN (missing values)
employee.dropna(subset=['Name', 'Age'], how = 'all', inplace = True, axis = 0)
#if how = any  --> either Name or Age columns
#if inplace = True --> permenantly remove the rows with missing values from the original dataframe, employee
# axis = 0 --> row wise operation
```
