```python
# Import the necessary libraries
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
# Load nbaallelo_log.csv into a dataframe
NBA = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/4510/2025-Fall-Lecture/CH7-Regression/data/nbaallelo_log.csv")
NBA.head(n=10)
```

```python
# Hot encode the game_result variable as a numeric variable with 0 for L and 1 for W
NBA.loc[NBA['game_result']=='L','game_result']=0
NBA.loc[NBA['game_result']=='W','game_result']=1
```

```python
# Store relevant columns as variables
X = NBA[['elo_i']].values.reshape(-1, 1)
y = NBA[['game_result']].values.reshape(-1, 1).astype(int)
```

```python
# Build the logistic model
logisticRegr = LogisticRegression()
model = logisticRegr.fit(X,np.ravel(y.astype(int)))
```
